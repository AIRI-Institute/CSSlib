```python
from csslib import CSS
```

**Set path to configuration file. Example of configurational file is given below.**

```python
{
  "result_dir": "CsPbI3",  # Directory with this name will be created at ./results directory. All results will be stored there.
  "structure_filename": "data/CsPbI3/CsPbI3.cif",  # Path to cif-file with initial structure description.
  "supercell": "2x1x1",  # Supercell size.
  "num_workers":  4,  # Number of workers that run Supercell software.
  "substitution": [  # A list with species substitution description.
    {
      "specie_to_substitute": "Pb",  # Specie to substitute (Pb).
      "substitute_with": "Cd",  # Abovementioned specie will be substituted with a given specie (Cd).
      "substitution_low_limit": 0.0,  # Substitution low limit (from 0.0 to 1.0). It is an atomic fraction of Cd in the resulting supercell.
      "substitution_high_limit": 0.05  # Substitution high limit (from 0.0 to 1.0). It is an atomic fraction of Cd in the resulting supercell.
    },
    {
      "specie_to_substitute": "I",
      "substitute_with": "Br",
      "substitution_low_limit": 0.0,
      "substitution_high_limit": 0.06
    }
  ]
}
```

```python
config_filename = "data/CsPbI3/config_CsPbI3.json"
css = CSS(config_filename)
```

**Read an initial structure from a cif-file.**

```python
css.read_structure()
```

**Generate substituted disordered structures with partial occupancies according to the configuration file. Save them to a cif-files at results/{result_dir}/supercell_input_cifs directory.**

```python
css.generate_substituted_disordered_structures()
```

**Run Supercell software to convert a crystallographic structures with partial occupancies and/or vacancies to ordinary supercell structures. Save them at results/{result_dir}/supercell_output directory.**

```python
css.run_supercell()
```

**Collect data from the output of Supercell software to a pandas dataframe and save it to a data.pkl.gz file at results/{result_dir} directory.**

```python
css.collect_data()
```

**Plot the group-subgroup graph using data.pkl.gz file at results/{result_dir} directory.**

```python
css.plot_group_subgroup_graph()
```

**Example of CSSlib usage in the case of 2D-material WSe2. There is simultaneoues substitution of W species with Mo and vacancy (represented by U species), Se species with S and vacancy (represented by Np species).**

```python
{
  "result_dir": "WSe2",  # Directory with this name will be created at ./results directory. All results will be stored there.
  "structure_filename": "data/Wse2/WSe2.cif",  # Path to cif-file with initial structure description.
  "supercell": "2x2x1",  # Supercell size. Composition of non-substituted resulting supercell will be W4Se8.
  "num_workers":  4,  # Number of workers that run Supercell software.
  "substitution": [  # A list with species substitution description.
    {
      "specie_to_substitute": "W",  # Specie to substitute (W).
      "substitute_with": "Mo",  # Abovementioned specie will be substituted with a given specie (Mo).
      "substitution_low_limit": 0.0,  # Substitution low limit (from 0.0 to 1.0). It is an atomic fraction of Mo in the resulting supercell.
      "substitution_high_limit": 0.3333333  # Substitution high limit (from 0.0 to 1.0). It is an atomic fraction of Mo in the resulting supercell.
                                            # I.e. W species will be substituted with Mo in range from 0 to 100 % of W content (W4Mo0..., W3Mo1..., W2Mo2..., W1Mo3..., W0Mo4...). 
    },
    {
      "specie_to_substitute": "W",  # Specie to substitute (W).
      "substitute_with": "U",  # W species will be substituted with U species. In this case U specie immitates vacancy in W-sublattice.
                               # This is a convenient way to do subsequent structure selection and collect statistics among built search space.
                               # However, one should remember to remove these species before DFT calculations, ML-model training, etc.
      "substitution_low_limit": 0.0,  # Substitution low limit (from 0.0 to 1.0). It is an atomic fraction of U (vacancy in W-sublattice) in the resulting supercell.
      "substitution_high_limit": 0.1666667  # Substitution high limit (from 0.0 to 1.0). It is an atomic fraction of U (vacancy in W-sublattice) in the resulting supercell.
                                            # I.e. W species will be substituted with vacancy in range from 0 to 50 % of W content (W4U0..., W3U1..., W2U2...).
    },
    {
      "specie_to_substitute": "Se",  # Specie to substitute (Se).
      "substitute_with": "S",  # Abovementioned specie will be substituted with a given specie (S).
      "substitution_low_limit": 0.0,  # Substitution low limit (from 0.0 to 1.0). It is an atomic fraction of S in the resulting supercell.
      "substitution_high_limit": 0.6666667  # Substitution high limit (from 0.0 to 1.0). It is an atomic fraction of S in the resulting supercell.
                                            # I.e. Se species will be substituted with S in range from 0 to 100 % of Se content (Se8S0..., ..., Se0S8...).
    },
    {
      "specie_to_substitute": "Se",  # Specie to substitute (Se).
      "substitute_with": "Np",  # Se species will be substituted with Np species. In this case Np specie immitates vacancy in Se-sublattice.
                                # This is a convenient way to do subsequent structure selection and collect statistics among built search space.
                                # However, one should remember to remove these species before DFT calculations, ML-model training, etc.
      "substitution_low_limit": 0.0,  # Substitution low limit (from 0.0 to 1.0). It is an atomic fraction of Np (vacancy in Se-sublattice) in the resulting supercell.
      "substitution_high_limit": 0.3333333  # Substitution high limit (from 0.0 to 1.0). It is an atomic fraction of Np (vacancy in Se-sublattice) in the resulting supercell.
                                            # I.e. Se species will be substituted with vacancy in range from 0 to 50 % of Se content (Se8Np0..., ..., Se4Np4...).
    }
  ]
}
```

```python
config_filename = "data/WSe2/config_WSe2.json"
css = CSS(config_filename)
```

```python
css.read_structure()
```

```python
css.generate_substituted_disordered_structures()
```

```python
css.run_supercell()
```

```python
css.collect_data()
```

```python
css.plot_group_subgroup_graph()
```
